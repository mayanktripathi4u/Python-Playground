```python
import mysql.connector
from mysql.connector import Error

def create_connection():
    try:
        connection = mysql.connector.connect(
            host='localhost',  # Replace with your host
            database='moneycontrol_flaskapp',   
            user='root',  # Replace with your MySQL username
            password='mysqlroot'  # Replace with your MySQL password
        )
        if connection.is_connected():
            print("Connection successful")
            return connection
    except Error as e:
        print(f"Error: '{e}'")
        return None
```

```python
def insert_data(connection, sql_stmt = None, df = None):
    cursor = connection.cursor()

    if sql_stmt:
        print("sql stamenet received")
        cursor.execute(sql_stmt)
    elif df:
        print("DataFrame received.")
        # Insert distinct categories
        for category in df['Category'].unique():
            print(f"Inserting Category : {category}")
            cursor.execute("INSERT IGNORE INTO category (name) VALUES (%s)", (category,))

        # Insert products and maintain category relationships
        for index, row in df.iterrows():
            print(f"Insert products and maintain category relationships : {row['Category']} --> {row['Product']}")
            # Get the category id
            cursor.execute("SELECT id FROM category WHERE name = %s", (row['Category'],))
            category_id = cursor.fetchone()[0]

            # Insert product (ignore duplicate products)
            cursor.execute("INSERT IGNORE INTO sub_category (name, category_id) VALUES (%s, %s)", (row['Product'], category_id))

            # Get the product id
            cursor.execute("SELECT id FROM sub_category WHERE name = %s", (row['Product'],))
            subcategory_id = cursor.fetchone()[0]

            # Insert item (ignore duplicate items)
            print(f"Insert item (ignore duplicate items) : {subcategory_id} --> {row['Item']}")
            cursor.execute("INSERT IGNORE INTO product (name, subcategory_id) VALUES (%s, %s)", (row['Item'], subcategory_id))

    # Commit the transaction
    connection.commit()
    cursor.close()

```

```python
# Insert the data
connection = create_connection()
sql_stmt = "INSERT IGNORE INTO category (name) VALUES ('Grocery')"
if connection:
    insert_data(connection, sql_stmt=sql_stmt)
    connection.close()
```

```python
def generate_sql_stmt(df):
    # Get unique Categories and Products and Items
    unique_categories = df['Category'].unique()
    unique_products = df[['Category', 'Product']].drop_duplicates()
    unique_items = df[['Category', 'Product', 'Item']].drop_duplicates()

    # Insert statements list
    insert_statements = []

    # Step 1: Insert Categories
    for category in unique_categories:
        insert_statement = f"INSERT IGNORE INTO category (name) VALUES ('{category}');"
        insert_statements.append(insert_statement)

    # Step 2: Insert Products linked to Categories
    for _, row in unique_products.iterrows():
        insert_statement = f"""
        INSERT IGNORE INTO sub_category (name, category_id) 
        VALUES ('{row['Product']}', 
        (SELECT id FROM category WHERE name = '{row['Category']}'));
        """ 
        insert_statements.append(insert_statement)

    # Step 3: Insert Items linked to Products
    for _, row in unique_items.iterrows():
        insert_statement = f"""
        INSERT IGNORE INTO product (name, subcategory_id) 
        VALUES ('{row['Item']}', 
        (SELECT id FROM sub_category WHERE name = '{row['Product']}'));
        """
        insert_statements.append(insert_statement)

    # Print the insert statements
    for statement in insert_statements:
        # print(statement)
        connection = create_connection()
        if connection:
            insert_data(connection, sql_stmt=statement)
            connection.close()
```
