# 5. Polymorphism
Problem Statement
- Demonstrate polymorphism by defining a method fuel_type in both Car and ELectricCar classes, but with different behaviors.

```python
class Car:
    def __init__(self, user_brand, user_model):
        self.__brand = user_brand
        self.model = user_model

    # get_<attribute name>()
    def get_brand(self):
        return self.__brand + " !"
 
    def full_name(self):
        return f"{self.__brand} - {self.model}"

    def fuel_type(self):
        return "Petrol or Diesel"

class ElectricCar(Car):
    def __init__(self, brand, model, battery_size):
        super().__init__(brand, model)

        self.battery_size = battery_size
        
    def fuel_type(self):
        return "Electric Charge"

tesla_car = ElectricCar("Tesla", "Model S", "85kWH")
print(tesla_car.fuel_type())

tata_car = Car("Tata", "Safari")
print(tata_car.fuel_type())
```

# 6. Class variables
Add a class variable to Car that keeps track of the number of cars created.

```python
class Car:
    total_car = 0
    def __init__(self, user_brand, user_model):
        self.__brand = user_brand
        self.model = user_model
        Car.total_car += 1

    # get_<attribute name>()
    def get_brand(self):
        return self.__brand + " !"
 
    def full_name(self):
        return f"{self.__brand} - {self.model}"

    def fuel_type(self):
        return "Petrol or Diesel"

class ElectricCar(Car):
    def __init__(self, brand, model, battery_size):
        super().__init__(brand, model)

        self.battery_size = battery_size
        
    def fuel_type(self):
        return "Electric Charge"

tesla_car = ElectricCar("Tesla", "Model S", "85kWH")
print(tesla_car.fuel_type())

tata_car = Car("Tata", "Safari")
print(tata_car.fuel_type())

nissan_car = Car("Nissan", "Altima")
print(nissan_car.fuel_type())

print(f"Total Cars so far: {Car.total_car}")
```

# 7. Static Method
Add a static method to the Car class that returns a general description of a Car.

These Static Method are also part of Decorators.

```python
class Car:
    total_car = 0
    def __init__(self, user_brand, user_model):
        self.__brand = user_brand
        self.model = user_model
        Car.total_car += 1

    # get_<attribute name>()
    def get_brand(self):
        return self.__brand + " !"
 
    def full_name(self):
        return f"{self.__brand} - {self.model}"

    def fuel_type(self):
        return "Petrol or Diesel"

    @staticmethod
    def general_description():
        return "Cars are means of transport."


nissan_car = Car("Nissan", "Altima")

# Since general_description is a static method, it cannot be accessed from object.
# print(nissan_car.general_description())

# Access using Class
print(Car.general_description())

```

# 8. Property Decorators
Use a property decorator in the Car class to make the model attribute read-only.

```python
class Car:
    total_car = 0
    def __init__(self, user_brand, user_model):
        self.__brand = user_brand
        self.model = user_model
        Car.total_car += 1

    # get_<attribute name>()
    def get_brand(self):
        return self.__brand + " !"
 
    def full_name(self):
        return f"{self.__brand} - {self.model}"

    def fuel_type(self):
        return "Petrol or Diesel"

    @staticmethod
    def general_description():
        return "Cars are means of transport."


nissan_car = Car("Nissan", "Altima")
print(f"Able to Modify the Model - {nissan_car.model}")
nissan_car.model = "City"
print(f"Model after modication - {nissan_car.model}")
```

```python
# Lets make Model as Read-Only.
class Car:
    total_car = 0
    def __init__(self, user_brand, user_model):
        self.__brand = user_brand
        self.__model = user_model
        Car.total_car += 1

    # get_<attribute name>()
    def get_brand(self):
        return self.__brand + " !"
 
    def full_name(self):
        return f"{self.__brand} - {self.__model}"

    def fuel_type(self):
        return "Petrol or Diesel"

    @staticmethod
    def general_description():
        return "Cars are means of transport."

    @property
    def model(self):
        return self.__model

nissan_car = Car("Nissan", "Altima")
print(f"Model - {nissan_car.model}")
# Not allowed to edit or update the model.
```

# 9. CLass Inheritance and isinstance() function
Demonstrate the use of isinstance() to check if my_tesla is an instance of Car and ElectricCar.

```python
class Car: 
    def __init__(self, user_brand, user_model):
        self.__brand = user_brand
        self.model = user_model


class ElectricCar(Car):
    def __init__(self, brand, model, battery_size):
        super().__init__(brand, model)

        self.battery_size = battery_size
        

tesla_car = ElectricCar("Tesla", "Model S", "85kWh")

print(isinstance(tesla_car, Car))
print(isinstance(tesla_car, ElectricCar))
```
