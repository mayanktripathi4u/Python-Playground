```python
import pandas as pd
import chromadb
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer
```

```python
# Load the DataSet
df = pd.read_csv("amazon.csv")
```

```python
df.head()
```

```python
df = df[["product_name", "category", "about_product"]]

df.head()
```

```python
df['combined_text'] = ""

for col in df.columns:
    # df['combined_text'] += df[col].astype(str) + " "
    df['combined_text'] += df[col] + "\n"

df.head(2)
```

```python
# Create Embeddings

# Load the embedding model
model = SentenceTransformer('BAAI/bge-small-en')

# Convert the text to embeddings
embeddings = model.encode(df['combined_text'].tolist(), show_progress_bar=True)
```

```python
# Initialize ChromaDB client
client = chromadb.Client(Settings())
```

```python
collection_name = "product_search"
collection = client.create_collection(name=collection_name)
```

```python
# Prepare the data for insertion
ids = []
metadata = []

for i in range(len(embeddings)):
    ids.append(str(i))  # Use string IDs
    metadata.append({
        "product_name": df['product_name'][i],
        "category": df['category'][i]
    })

# Insert the embeddings into the collection
collection.add(
    ids=ids,
    embeddings=embeddings,
    metadatas=metadata,
    # documents=df['combined_text'].tolist()
)


# # Save the collection to disk
# collection.persist()
```

```python
# Create function for searching / retrieving products
def vector_search(query, model, collection, top_n = 10):
    # Get the embedding for the query
    query_embedding = model.encode([query])

    # Search the Chroma Index
    result = collection.query(
        query_embeddings=query_embedding,
        n_results=top_n
    )
    return result
```

```python
# Sample query
query = "wireless headphones with noise cancellation"

# Perform the search
results = vector_search(query, model, collection, 5)
```
